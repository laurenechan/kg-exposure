```python
# import required module
import os, csv, re
import pandas as pd
# assign directory
directory = '/Users/laurenchan/Desktop/GIT/kg-exposure/USDA ACUP/'
# iterate over files in
# that directory
commodityDict = {}
for filename in os.listdir(directory):
    if not filename.endswith(".csv"):
        continue
    f = os.path.join(directory, filename)
    with open(f, 'r', encoding='utf8') as inf:
        print(filename)
        rd = csv.reader(inf, delimiter=',')
        c=0
        domainsList = []
        for row in rd:
            if c ==0:
#                 print(row)
                commIndex = row.index('Commodity')
                domainIndex = row.index('Domain Category')
                yearIndex = row.index('Year')
#                 print(commIndex, domainIndex)
            else:
                if 2016 <= int(row[yearIndex]) <= 2020:
                    domainsList.append(row[domainIndex])
            c+=1
        commodity=row[commIndex]
        commodityDict[commodity] = set(domainsList)
        pattern = re.compile("[a-zA-Z, ]+: \((.+) = \d+\)")
        edges = []
        for key, value in commodityDict.items():
            for set_item in value:
                result = pattern.search(set_item)
                if result == None:
                    continue
                chemical = result.group(1)
                edges.append({'id': key, 'text': chemical})
        df = pd.DataFrame(edges)
        df.to_csv('fruit_fungicides_2016_20.tsv', sep='\t')
        #print(commodityDict)  
        
        
        
        
        #Is there a way I can get this into a csv that has one column of which commodity and the second of which chemical? 
```
